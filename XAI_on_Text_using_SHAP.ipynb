```python
import datasets
import numpy as np
import transformers
import shap
```

```python
dataset = datasets.load_dataset("imdb", split="test")
```

```python
short_data = [v[:500] for v in dataset["text"][:20]]
```

```python
print(dataset[0])
```

```python
print(short_data[0])
```

```python
classifier = transformers.pipeline("sentiment-analysis", return_all_scores=True)
```

```python
text=["I absofuckinglutely LOVED the movie. A fun watch!", "I hated the ending of the movie but overall it was good decent movie great one time watch"]

result=classifier(text)

print(result)
```

```python
print("1. ", short_data[0], "\n")
print("2. ", short_data[1], "\n")
print("3. ", short_data[2], "\n")
print("\n")
print(classifier(short_data[:3]))
```

```python
from sklearn.metrics import accuracy_score

predictions = classifier(short_data)
predicted_labels = []
for pred in predictions:
  predicted_label = pred['label']
  predicted_labels.append(0 if predicted_label == 'NEGATIVE' else 1)

true_labels = dataset["label"][:20]

accuracy = accuracy_score(true_labels, predicted_labels)
print(f"Accuracy: {accuracy}")
```

```python
print(true_labels)
```

```python
explainer = shap.Explainer(classifier)
```

```python
shap_values = explainer(short_data[0])
```
